```python
df['dt_inicial'] = [datetime(df['Año inicial de observación'][ii],df['Mes inicial de observación'][ii],df['Día inicial de observación'][ii]) for ii in range(len(df))]
df['dt_final'] = [df['dt_inicial'][ii] + timedelta(days=int(df['Duración total del ciclo (en días)'][ii])) for ii in range(len(df))]
```

### Generando la tabla

```python
df_completo = pd.DataFrame(columns=['Estaciones', 'Años', 'Meses', 'Días', 'Cultivos', 'Identificación de los campos', 'Cantidad de Campos', 'Cantidad de observaciones', 'Fases del cultivo'])
df_completo = df_completo.astype({'Estaciones': str, 'Años': int, 'Meses': int, 'Días': int, 'Cultivos': str, 'Identificación de los campos': int, 'Cantidad de Campos': int, 'Cantidad de observaciones': int, 'Fases del cultivo': str})
df_mes = pd.DataFrame(columns=['Estaciones', 'Años', 'Meses', 'Cultivos', 'Identificación de los campos', 'Cantidad de Campos', 'Cantidad de observaciones', 'Fases del cultivo'])

for ii in range(len(df)):
    df_dia_empty = pd.DataFrame(columns=['Estaciones', 'Años', 'Meses', 'Días', 'Cultivos', 'Identificación de los campos', 'Cantidad de Campos', 'Cantidad de observaciones', 'Fases del cultivo'])
    df_dia_empty = df_dia_empty.astype({'Estaciones': str, 'Años': int, 'Meses': int, 'Días': int, 'Cultivos': str, 'Identificación de los campos': int, 'Cantidad de Campos': int, 'Cantidad de observaciones': int, 'Fases del cultivo': str})
    lista = calculador_de_fecha(df.loc[ii, 'dt_inicial'], df.loc[ii, 'dt_final'], df.loc[ii, 'Frecuencia de observación'])

    # df_dia_empty
    df_dia_empty['Años'] = list(map(year, lista))
    df_dia_empty['Meses'] = list(map(month, lista))
    df_dia_empty['Días'] = list(map(day, lista))
    df_dia_empty['Estaciones'] = df.loc[ii, 'Estaciones']
    df_dia_empty['Cultivos'] = df.loc[ii, 'Cultivos']
    df_dia_empty['Fases del cultivo'] = df.loc[ii, 'Fases del cultivo']
    df_dia_empty['Identificación de los campos'] = int(df.loc[ii, 'Cultivos'].split()[-1])
    df_dia_empty['Cantidad de Campos'] = 1
    df_dia_empty['Cantidad de observaciones'] = 1

    df_completo = pd.concat([df_completo, df_dia_empty])
df_dia = df_completo.loc[:, ['Estaciones', 'Años', 'Meses', 'Días', 'Cultivos', 'Fases del cultivo']]
df_mes = df_completo.loc[:, ['Estaciones', 'Años', 'Meses', 'Cultivos', 'Identificación de los campos', 'Cantidad de Campos', 'Cantidad de observaciones', 'Fases del cultivo']].groupby(['Estaciones', 'Años', 'Meses', 'Cultivos']).count()
df_mes['Identificación de los campos'] = [int(df_mes.index.get_level_values(3)[ii].split()[-1]) for ii in range(len(df_mes))]
df_mes['Cantidad de Campos']  = 1
df_mes['Fases del cultivo'] = '1'
df_mes['Fases del cultivo'] = df_completo.loc[:,['Estaciones', 'Años', 'Meses', 'Cultivos', 'Fases del cultivo']].groupby(['Estaciones', 'Años', 'Meses', 'Cultivos']).agg(', '.join)
df_mes['Fases del cultivo'] = df_mes['Fases del cultivo'].apply(lambda x: ', '.join(list(set(x.split(', ')))))
df_mes.reset_index(inplace=True)
```
